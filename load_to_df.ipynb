```python
import pandas as pd
import numpy as np
from scipy import ndimage #ndimage provides a large amount of functions for image processing: https://docs.scipy.org/doc/scipy/reference/ndimage.html 
from skimage import color #many tutorials: https://github.com/scikit-image/skimage-tutorials/tree/master/lectures
import matplotlib.pyplot as plt # matplotlib is the major lib for plotting
import imageio
from PIL import Image
import os
```

```python

# assign searchterm directory
base = 'body lotion advertisement'
directory = 'photos/' + base #photos/cosmetic advertisement, photos/hair product advertisement, photos/nail polish advertisement, photos/shampoo advertisement
directory_out = 'photos_jpeg/' + base
# iterate over files in
# that directory
for filename in os.listdir(directory):
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.png'):
        try:
            img = Image.open(os.path.join(directory, filename)) # open the image file
            img.verify() # verify that it is, in fact an image
        except (IOError, SyntaxError) as e:
            print('Bad file:', filename)
            os.remove(os.path.join(directory, filename)) 
    
```

```python
#convert all images to jpeg and load it into new directory
for filename in os.listdir(directory):
    if filename == '.DS_Store':
        continue
    else:
        f_nam = f'{str.split(filename, ".")[0]}.jpeg'
        img = Image.open(os.path.join(directory, filename))
        im  = img.convert('RGB')
        print(os.path.join(directory_out, f_nam))
        im.save(os.path.join(directory_out, f_nam))
```

```python
#get shapes of images into dataframe
im_y = []
im_x = []
im_filename = []
for filename in os.listdir(directory_out):
    #read image
    try:
        im = imageio.imread(os.path.join(directory_out, filename))
        im_y.append(im.shape[0])
        im_x.append(im.shape[1])
        im_filename.append(str.split(filename, '.')[0])
    except:
        print(f'{filename} failed')
        
im_dic = {'Width':im_x, 'Height': im_y, 'Filename': im_filename, 'Folder': str.split(directory_out, '/')[1]}
df_im = pd.DataFrame(im_dic)
df_im['aspect_ratio'] = df_im['Height']/df_im['Width']
```

(y, x, channels)

```python
#round aspect ratio
df_im['aspect_ratio_round'] = round(df_im['aspect_ratio'], 1)
```

```python
#check occurences of aspect ratios
df_im.groupby('aspect_ratio_round').count()
```

```python
#export dataframe to csv
out_name = str.split(directory_out, '/')[-1]
out_path = f'DATA/{out_name.replace(" ", "_")}.csv' 
df_im.to_csv(out_path)
```
